```python
#Creating a set of non-linear data
import numpy as np

np.random.seed(40)
m = 1000
X = 10 * np.random.rand(m, 1) - 5
y = 2 * X + X**2 + np.random.randn(m, 1)
```

```python
# plotting the data
import matplotlib.pyplot as plt

plt.plot(X, y, "b*")
plt.xlabel("x")
plt.ylabel("y",rotation = 1)
plt.show()
```

```python
# transforming the data to polynomial features of degree 4
from sklearn.preprocessing import PolynomialFeatures

poly_features = PolynomialFeatures(degree=4, include_bias=False)
x_poly = poly_features.fit_transform(X)
```

```python
# Lasso regularizer with alpha 0.01
from sklearn.linear_model import Lasso

lasso = Lasso(0.01)
lasso.fit(x_poly, y);
```

```python
# making predictions and displaying the data and the results
y_pred = lasso.predict(x_poly)
plt.plot(X, y, "b*", label="Ground Truth")
plt.plot(X, y_pred, "r.", label="Predicted")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()    
```

```python
lasso.coef_
```

#### Comments:

We observe that the model retrieves the factors of the first and
second-order features while eliminating the contribution of the third and fourth-order features.

```python
#Ridge regularizer with alpha 0.01
from sklearn.linear_model import Ridge

ridge = Ridge(0.01)
ridge.fit(x_poly, y);
```

```python
#making predictions and displaying the data and the results
y_pred = ridge.predict(x_poly)
plt.plot(X, y, "b*", label="Ground Truth")
plt.plot(X, y_pred, "r.", label="Predicted")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```

```python
ridge.coef_
```

#### Comments:

Like the Lasso model, we observe that the Ridge one retrieves the factors of the first and second-order features while
eliminating the contribution of the third and fourth-order features.
